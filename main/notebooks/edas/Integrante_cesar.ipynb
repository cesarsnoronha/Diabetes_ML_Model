```python
corr = df.corr()

import numpy as np

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.savefig('../../outputs/imgs/corr_plot.png', bbox_inches='tight')
```

Correlações:
A primeira coluna da tabela de correlação podemos a correlação entre a presentão de diabetes e cada outra variável. Entre as correlações negativas temos: PhysActivity, Fruits, Veggies e HvyAlcoholConsump. Para as 3 primeiras variáveis temos essa corelação pois elas estão relacionadas a formas de evitar a diabetes. Para o caso so alto consumo de alcohol, acredito que temos uma correlação negativa pois pessoas não saldáveis tem menos condições de consumir alcohol. As outras variáveis tem correlaçõa positiva.

```python
corr = df.corr()
corr2 =  corr.sort_values('Diabetes_binary',axis=1,ascending = False)
corr2 =  corr2.sort_values('Diabetes_binary',ascending = False)

import numpy as np

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr2, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.savefig('../../outputs/imgs/corr_plot.png', bbox_inches='tight')
```

Correlações organizadas:
Gráficos de correlações por ordem de Correlação com a presença diabetes. Nesse gráfico podemos observar grupos de variáveis de correlação positiva e grupos de correlação negativa com a presença de diabetes. Cada grupo tende a ter uma correlação positiva entre sí e negativa com o outro grupo. Também podemos observar algumas excessões como se a pessoa fuma, e consome muito alcool. Pressão Saguínea alta:
Entre os sujeitos Com uma taxa de pressão alta (HighBP = 1) temos uma maior prevalencia de sujeitos com diabetes, enquanto para os sujeitos sem pressão alta (HighB = 0) temos uma maior prevalência de sujeitos sem diabetes.

```python
fig, ax = plt.subplots(figsize = ( 5 , 3 ))
sns.countplot(ax = ax, x='Education', hue='Diabetes_binary', data=df);

# Set label for x-axis
ax.set_xlabel( "Educação" , size = 12 )
  
# Set label for y-axis
ax.set_ylabel( "Quantidade" , size = 12 )
  
# Set title for plot
ax.set_title( "Distribuição da proporção de diabetes por educação" , size = 10 )
  
# Display figure

plt.savefig('../../outputs/imgs/bar_plot_education.png', bbox_inches='tight')
plt.show()
```

Nível de educação:
Até um grau de educação 5 temos uma maior proporção de sujeitos com diabetes do que sem diabetes, com excessão do grau de educação 6 que temos sujeitos com menos diabetes. Com grau de educação 5 podemos observar uma tendência de redução.


```python
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(12, 12))
plt.subplots_adjust(hspace=0.5 ,wspace=0.3)
fig.suptitle("gráficos as variáveis com correlação positiva com diabetes pela presença de Diabetes", fontsize=15, y=0.95)

PositivaCorr = [ 'HighBP', 'HighChol', 'HeartDiseaseorAttack', 'GenHlth','PhysHlth', 'DiffWalk', 'BMI', 'Age']



for column, ax in zip(PositivaCorr, axs.ravel()):
    if column == 'PhysHlth':
       df[column].value_counts().plot.pie(ax=ax)
       ax.set_title( f"Distribuição de {column}" , size = 10 )
    elif column == 'BMI':
        sns.countplot(ax = ax, x=column, hue='Diabetes_binary', data=df);
        ax.set_title( f"Distribuição da proporção de diabetes por {column}" , size = 10 )
        ax.set_xticklabels([])
    else:
        sns.countplot(ax = ax, x=column, hue='Diabetes_binary', data=df);
        ax.set_title( f"Distribuição da proporção de diabetes por {column}" , size = 10 )
    # Set label for x-axis
   
    
plt.savefig('../../outputs/imgs/bar_plot_positive_corr.png')    
plt.show()
```

Variáveis de correlação positiva:
Podemos observar a a correlação da presença de diabetes em relação a alta pressão sanguínea (HighBP), altas taxas de colesterol (HighChol), se o sujeito ja teve parada cardíaca (HearthDiseaseorAttack), e se o sujeito tem dificuldade de andar.  
Também podemos ver a que a distribuição da massa corporea e a idade dos sujeitos com  diabetes estão a direita da curva do que os sem diabetes, indicando a diabetes está relacionada a uma maior massa corporea e a uma idade mais avançada. 


```python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))
plt.subplots_adjust(hspace=0.5 ,wspace=0.3)
fig.suptitle("gráficos as variáveis com correlação negativa com diabetes pela presença de Diabetes", fontsize=15, y=0.95)

NegativaCorr = [ 'PhysActivity', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'Education', 'Income']

for column, ax in zip(NegativaCorr, axs.ravel()):
    
    sns.countplot(ax = ax, x=column, hue='Diabetes_binary', data=df);

    # Set label for x-axis
    ax.set_title( f"Distribuição da proporção de diabetes por {column}" , size = 10 )
    
plt.savefig('../../outputs/imgs/bar_plot_negative_corr.png')     
plt.show()
```

Variáveis de correlação negativa:
para as variáveis com correlação negativa a presença de diabetes podemos podemos observar uma relação mais fraca do que as variáveis de correlação positiva. Para as variáveis Educação (Education) e Renda (Income) podemos observar uma relação mais forte para os sujeitos de maior renda ou educação.
